```python

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split 
from sklearn.linear_model import LogisticRegression
%matplotlib inline
```

# loading the training data

```python
data=pd.read_csv("train.csv")
data.head()
```

# loading the test data

```python
test=pd.read_csv("test.csv")
test.head()
```

# preprocessing

```python
data=data.drop(["id"], 1)

#deleting the first column as the id shouldn't affect the learining
```

```python
color=data["color"]
goal=data["type"]

#preparing two dataframes to change categorical data
```

```python
from sklearn import preprocessing
le1 = preprocessing.LabelEncoder()
le1.fit(color)
color=le1.transform(color) 
from sklearn import preprocessing
le2 = preprocessing.LabelEncoder()
le2.fit(goal)
goal=le2.transform(goal) 

# using label encoder to replace categorical data with integral value
```

```python
data=data.drop(["type","color"], 1)
data.head()

# dropping the old columns type and color 
```

```python
color = pd.DataFrame(color,columns=['color'])
goal = pd.DataFrame(goal,columns=['goal'])

# label encoder return a different data type
# so i have to return them to dataframe 
```

```python
data = pd.concat([data,color,goal],axis=1)
data.head()

# now i will concatenate all the data frame 
# so that this is the training data 
# ready and prepared to be split
```

```python
X = data.ix[:,(0,1,2,3,4)].values
y = data.ix[:,-1].values

# spliting the data to features and goals
```

# fitting the data

```python
LogReg = LogisticRegression()
LogReg.fit(X, y)

# makeing the algorithm and fitting the data
```

# preparing the test set

```python
test.head()
```

```python
color=test["color"]

from sklearn import preprocessing
le3 = preprocessing.LabelEncoder()
le3.fit(color)
color=le3.transform(color) 

# using label encoder to replace categorical data with integral value
```

```python
test=test.drop(["color"], 1)
test.head()

# dropping the old columns type and color 
```

```python
color = pd.DataFrame(color,columns=['color'])

# label encoder return a different data type
# so i have to return them to dataframe 
```

```python
test = pd.concat([test,color],axis=1)
test.head()

# now i will concatenate all the dataframes 
# so that this is the test data
# is ready to make pridection
```

```python
yhat = test.ix[:,1:].values
y_pred = LogReg.predict(yhat)

# getting the pridections
```

```python
y_pred

#the pridectio as labels
```

```python
y=le2.inverse_transform(y_pred)

# replacing labels with actual names
```

```python
y
```

# preparing the pridections to be saved as csv file

```python
ids = test.ix[:,0].values
goal = pd.DataFrame(y,columns=['type'] ) 
ids = pd.DataFrame(ids,columns=['id'] )  

# transforming numpy arrays to datafram to be saved
```

```python
out=pd.concat([ids, goal],axis=1)

#concatenate the two data frames
```

```python
out
```

```python
out.to_csv('out.csv',index = False)

# saving the pridections as csv file
```
